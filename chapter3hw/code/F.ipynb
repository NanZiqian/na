```python
import matplotlib.pyplot as plt
import numpy as np

# input:
# t stores the divided points, expect to calculate [ti-1,...,ti+n]
# x = linspace(expected interval)
# return a vector of [ti-1,...,ti+n](t-x)_+^n
def divided_difference(t,x,n):
    N = len(t)
    if N == 1:
        return (t[0]-x)**n*((t[0]-x)>0)
    else:
        return (divided_difference(t[1:],x,n) - divided_difference(t[:-1],x,n))/(t[-1]-t[0])
    
for n in range(0,2):
    t = np.linspace(1,n+2,n+2)
    x = np.linspace(0,n+3,100)
    
    plt.figure(n)
    for j in range(n+2):
        for i in range(j,n+2):
            plt.subplot(n+2,n+2,i*(n+2)+j+1)
            y = divided_difference(t[i-j:i+1],x,n)
            #plt.xlim(0,n+3)
            #plt.ylim(-0.1,n+1)
            plt.plot(x,y)

```
