# B

$UFL(F) = \beta^L$

$OFL(F) = (\beta-\beta^{-p})*\beta^U$

Cardinality of $F = (\beta)^{p+1}(U-L+1)+1$

```python
beta = 2
p = 2
L = -1
U = 1

UFL = beta**L
OFL = (beta-beta**(-p))*beta**U
print("UFL = ",UFL, "; OFL = ",OFL,"\n")

points = []
for ii in range(L,U+1):
    points.append(np.linspace(1,beta-beta**(-p),beta**p)*beta**ii)

print("Points in F = ",points,"\n")
plt.gca().get_yaxis().set_visible(False)
color = ['r','g','b','y','k','c','m']
plt.figure(1)
for ii in range(len(points)):
    for i in range(len(points[ii])):
        plt.plot([points[ii][i],points[ii][i]],[0,1],color=color[ii])
    for i in range(len(points[ii])):
        plt.plot([-points[ii][i],-points[ii][i]],[0,1],color=color[ii])
plt.show()

plt.figure(2)
plt.gca().get_yaxis().set_visible(False)
points.append(np.linspace(0,1-beta**(-p),beta**p)*beta**L)
print("Points in extended F = ",points,"\n")
for ii in range(len(points)):
    for i in range(len(points[ii])):
        plt.plot([points[ii][i],points[ii][i]],[0,1],color=color[ii])
    for i in range(len(points[ii])):
        plt.plot([-points[ii][i],-points[ii][i]],[0,1],color=color[ii])
plt.show()
```
