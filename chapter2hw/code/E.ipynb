```python
import matplotlib.pyplot as plt
import numpy as np

txt = open("output/E.txt",'r')
parameters = [[None]]*2
vector_x = [0,6,10,13,17,20,28]

PNum = 0
for lines in txt:
    parameters[PNum] = lines.split()
    PNum += 1

def estimate(n,vector_x,parameters,x):
    result = float(parameters[n])
    #print(result)
    for i in range(n,0,-1):
        result *= (x-vector_x[i-1])
        result += float(parameters[i-1])
    return result
```

```python
for problem in range(2):
    # n points interpolation
    n = 7
    p = np.zeros(1000);
    x = np.linspace(0, 30, 1000)
    for i in range(0,1000):
        p[i] = estimate(n-1,vector_x,parameters[problem],x[i])
    plt.figure(problem)
    plt.plot(x, p, 'r')
    plt.xlabel('x')
    plt.ylabel('p')
    plt.title('average weight of larvae')
    plt.legend(['estiamted curve'])
    plt.savefig('output/E_'+str(problem)+'.png')
```
