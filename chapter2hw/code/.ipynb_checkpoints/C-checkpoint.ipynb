```python
import matplotlib.pyplot as plt
import numpy as np

txt = open("output/C.txt",'r')
parameters = [[None]]*4

PNum = 0
for lines in txt:
    parameters[PNum] = lines.split()
    PNum += 1

INum = [None]*4
for i in range(4):
    INum[i] = int(parameters[i][0])
    parameters[i] = parameters[i][1:]
# define estimate function
# n+1 points interpolation
# x -> p(x)
def estimate(n,vector_x,parameters,x):
    result = float(parameters[n])
    #print(result)
    for i in range(n,0,-1):
        result *= (x-vector_x[i-1])
        result += float(parameters[i-1])
    return result

def Chebyshev(n):
    result = [None]*n
    for i in range(n):
        result[i] = np.cos((2*i+1)*np.pi/(2*n))
    return result
```

```python
problem = 0
for n in INum:
    n = int(n)
    # n points interpolation
    vector_x = Chebyshev(n)
    p = np.zeros(1000);
    # plot the polynomials and the exact function
    x = np.linspace(-1, 1, 1000)
    f = 1/(1+25*x**2)
    for i in range(0,1000):
        p[i] = estimate(n-1,vector_x,parameters[problem],x[i])
    plt.figure(problem)
    plt.plot(x, f, 'b', x, p, 'r')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('n = '+str(n))
    plt.legend(['exact', 'interpolation'])
    plt.savefig('output/C_n'+str(n)+'.png')
    problem += 1
```
